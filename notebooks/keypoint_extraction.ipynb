```python
import cv2
import mediapipe as mp
import numpy as np
import os

# Initialize MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5)

def extract_keypoints(video_path, output_file):
    cap = cv2.VideoCapture(video_path)
    keypoints_sequence = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Convert frame to RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = pose.process(frame_rgb)

        if results.pose_landmarks:
            keypoints = []
            for landmark in results.pose_landmarks.landmark:
                keypoints.extend([landmark.x, landmark.y, landmark.z, landmark.visibility])
            keypoints_sequence.append(keypoints)

    cap.release()
    # Save keypoints as numpy array
    np.save(output_file, np.array(keypoints_sequence))

# Define paths
base_input_path = '/content/drive/MyDrive/dataset'
base_output_path = '/content/drive/MyDrive/keypoints'

# Process all videos
for category in ['fall', 'non_fall']:
    input_dir = f'{base_input_path}/{category}'
    output_dir = f'{base_output_path}/{category}'
    os.makedirs(output_dir, exist_ok=True)
    for video_file in os.listdir(input_dir):
        if video_file.endswith('.mp4'):  # Ensure it’s a video file
            video_path = os.path.join(input_dir, video_file)
            output_file = os.path.join(output_dir, f'{video_file.split(".")[0]}.npy')
            print(f"Processing {video_path}...")
            extract_keypoints(video_path, output_file)

pose.close()
```
